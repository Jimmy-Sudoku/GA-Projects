## Literature Review

### Classification vs Anomaly detection

To understand what types of machine learning models to use, we first have to understand what type of problem it is. 

Fraud detection is a process that detects and prevent fraudsters from getting something fraudulently ([sources](https://www.spiceworks.com/it-security/vulnerability-management/articles/what-is-fraud-detection/)). To detect fraudsters, it would need to differeitiate normal activites from fraud activities. To do that, it will first need to classify normal activities and fraud activites activities. Then distinguish these two activites correctly. There are two machine learning methods to do that and they are classification and anomaly detection. 


Classification is a supervised machine learning method that tries to classify observations into two or more labels. It also assumes that the class labels to be balanced. The reason for balanced class labels is important so that it will have equal importance.([source](https://medium.com/@ljyds/differences-between-classification-and-anomaly-detection-f0b4b4b990e2))

Anomaly detection, also known as outlier detection, objective is to differeitiate anomaly from the obsrvation. This means that it tries to classify observiations into two labels, normal and anomaly. It also implies that the class labels is imbalanced. There are three different types of anomaly detection which are supervised, semi-supervised and unsupervised. 
The difference between these three types is that all labels are provided for supervised anomaly detection, partial labels for semi-supervised and no labels provided for unsupervised. ([source](https://medium.com/@ljyds/differences-between-classification-and-anomaly-detection-f0b4b4b990e2))

Below are the comparasion table between the two machine learning methods. ([source](https://medium.com/@ljyds/differences-between-classification-and-anomaly-detection-f0b4b4b990e2))

|Machine Learning Method|Category|Dataset|Class|
|---|---|---|---|
|**Classification**|Supervised|Balanced|Binary<br> Multi-Class| 
|**Anomaly detection**|Supervised<br> Semi-Supervised<br> Unsupervised|Imbalanced|Binary| 

From this , depending on our datset, I will need to choose between the two machine learning problems. For this project, I first assumed that it is a supervised anomaly detection problem and because of that, I will dwell more indepth into supervised anaomaly detection in the literature review.

Simple imputer is a method of replacing missing values with either median, mean or mode. This process is usually done after train test split ([source]( https://stats.stackexchange.com/questions/95083/imputation-before-or-after-splitting-into-train-and-test/)) to prevent data leakage. ### Cross-Validation Method : Stratified K-Fold vs HoldOut

Cross-validation is a statistical resampling method that uses parts of the data to train and test a model on different iterations. The prupose of using cross-validation is to estimate the machine learning model's performance by predicting and estiamting how the model will perform in practice ([source](https://en.wikipedia.org/wiki/Cross-validation_(statistics))). Examples of its applications includes train test split and cross validation scores. There are a few different cross-validation methods but for this project, we will be focusing on two methods which are stratified k-fold and holdout method.

Holdout method is the process of separating the data into different parts for the modeling. Usually, the parts are split for training, validating and testing of models. For this reason, the hold-out method is usually used for both model evaluation and selection. The advantages of hold-out methods is that it is easy to understand, fast to implement and excute. The disadvantage is that it is not suitable for data that is imbalanced. The reason for that is that it is not generalised well. Meaning that there is a possiblility that some parts of the data might not have certain type class and this will affect the prediction score. To solve that issue, [sklearn has a parameters that allows the unbalanced data to be split in a stratified fashion](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html). Besides that, the parameters ,shuffle, will be set to True. The purpose is mixed the data before train test split so that the data which are sorted will be randomly assigned ([source](https://stats.stackexchange.com/questions/467816/what-is-the-advantage-of-shuffling-data-in-train-test-split)).


Stratified k-fold is the enhanced process of K-Fold cross-validation. K-fold cross-validation is the process where the whole data is splits into K-parts of equal size where each part is called K-fold. The difference between the stratified k-fold and normal k-fold lies in the word stratified. Meaning that stratified k-fold has additional parameter, which is having the same ratio of the target variable as the whole dataset. It is used for imbalanced dataset as the target varialbe is imbalanced, which means that the raio of the target variable will be the same throughout each split ([source](https://www.analyticsvidhya.com/blog/2021/11/top-7-cross-validation-techniques-with-python-code/)). Therefore, this will be used in the later part for cross-vaildation. 

### Feature Selection : Filter & Embedded Method (Algorithm Based Approach)

Feature selection is part of a processing step to choose the revelent features for the model to run. The reasons for doing feature selection are the following below:
- increases the traning speed of model faster as there are lesser datas to run and hence,
- easier to interpret the model
- improves the score of the model and
- reduces overfitting
([source](https://www.analyticsvidhya.com/blog/2016/12/introduction-to-feature-selection-methods-with-an-example-or-how-to-select-the-right-variables/#:~:text=Embedded%20methods%20combine%20the%20qualities,penalization%20functions%20to%20reduce%20overfitting.))

There are many methods of doing feature selection but our focus will be embedded and it will be used for this project.

Embedded method is the use of both fliter and wapper method.

Filter method is the selection of features that uses the properties of the features such as correlation coefficient and domain knowledge of the features([source](https://www.analyticsvidhya.com/blog/2020/10/feature-selection-techniques-in-machine-learning/)). The benefits of using filter method is that it is faster and does not requires alot of computation efforts to determine if the feature is important for model.

Wapper method is the use of machine learning model to train a subset of features and see how is it how it performs. Based on the previous model, it will choose to remove or add certran features. The benefit of wapper method is that the model itself will help us choose the features. However, it is very time consuming and requires hugh energy consumption and exensive computer to run ([source](https://www.analyticsvidhya.com/blog/2016/12/introduction-to-feature-selection-methods-with-an-example-or-how-to-select-the-right-variables/#:~:text=Embedded%20methods%20combine%20the%20qualities,penalization%20functions%20to%20reduce%20overfitting.)). 

Which is why, by using embedded method, which is the combined use of filter and wapper method, will be the most cost efficinet way of getting things done.

For filter method, correlation coefficient will be used and for embedded method, xgboost model will be used. The reason for using xgboost model for embedded method is that it uses algorithm based approach to pick the best feature importance ([source](https://medium.com/analytics-vidhya/feature-selection-for-dimensionality-reduction-embedded-method-e05c74014aa)).

### Recall vs Accuracy Score

Recall score is a measure of how much the model correctly identifies the True Positives ([source](https://www.analyticsvidhya.com/blog/2020/09/precision-recall-machine-learning/)). 

Accuracy score is a measure of how much the model correctly made the correct predictions ([source](https://deepchecks.com/how-to-check-the-accuracy-of-your-machine-learning-model/#:~:text=Accuracy%20score%20in%20machine%20learning,the%20total%20number%20of%20predictions.)). 

For anomaly problem, it is commonly to use recall score instead of accuracy as recall score shows the proportion of true anomalies was identified ([source](https://towardsdatascience.com/anomaly-detection-how-to-tell-good-performance-from-bad-b57116d71a10)).

Therefore, recall score will be used in this project.

### PR Curve vs ROC Curve

Receiver Operating Characteristic (ROC) curve is an analytical method used for predicting the probability of a class label ([source](https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/)).

Precision-Recall (PR) curve is a plot of the precision and recall in y-axis and x-axis respectfully. It is used to plot for different thresholds ([source](https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/)).

One of the difference between ROC curve and PR curve is that the ROC curve focuses more on accuracy score and PR curve focus on the precision and recall scores. This also means that ROC curves are better when there is a balance between each class and PR curve are better for imbalanced datasets ([source](https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/)). 

Besides that, for imbalance dataset, PR curve is said to have more information as compared to ROC curve ([source](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4349800/)). 

The below shows a diagram taken from ([source](https://datascience.stackexchange.com/questions/61467/clarification-on-train-test-and-val-and-how-to-use-implement-it)) :

## Preprocessing, Modeling & Postprocessing

### Data Preparation : Train - Val - Test - Split

```python
#Prepare data for train test split
X=df1.drop("flag",axis=1)

y=df1[["flag"]]

#Show the shape of the X and y features

display("The following is the shape of X : ",X.shape);display("The following is the shape of y :",y.shape)
```

```python
#Split dataset into train valid test 

#first split into train and remaining dataset
X_train,X_rem,y_train,y_rem =train_test_split(X,y,random_state=888,
                                               train_size=0.8,test_size=0.2)

#Second split into remaining dataset into valid and test dataset
X_valid,X_test,y_valid,y_test =train_test_split(X_rem,y_rem,
                                                 random_state=888,
                                                 train_size=0.5,test_size=0.5)

display("The following data for training are : ")
print(X_train.shape) 
print(y_train.shape)
display("The following data for validation are : ")
print(X_valid.shape)
print(y_valid.shape)
display("The following data for testing are : ")
print(X_test.shape)
print(y_test.shape)
```

### Stratified_kfold

```python
#for cross-val fold
stratified_kfold = StratifiedKFold(n_splits=10,shuffle=True)
#since the data is not time related, shuffle is set to true to reduce overfitting.
```

Since the data is imbalance, stratified k-fold cross validation is used.

### Feature Selection using Embedded Method  (Xgboost)

#### Model set-up

```python
xgboost_pipe=imbpipeline(steps=[["simpleimputer",SimpleImputer()],
                                ["smote",SMOTE()],
                                ['ss',StandardScaler()],
                                ["xgboost_classifier",XGBClassifier()]])
```

```python
xgboost_pipe.get_params()
```

```python
#format : name of pipeline __ actual terms
xgboost_pipe_params = [{'verbose': [False],
                        'smote__n_jobs': [-1],#to use all cpu cores
                        'smote__random_state': [888],#set a number so that the result is repliicable
                        'simpleimputer__strategy': ['most_frequent'],
                        'xgboost_classifier__random_state': [888],
                        'xgboost_classifier__n_jobs': [-1],
                        }]
```

```python
xgboost_pipe_gridsearch = GridSearchCV(xgboost_pipe,
                                       xgboost_pipe_params,
                                       cv=stratified_kfold,
                                       verbose=0,
                                       scoring="recall",
                                       n_jobs=-1,)
```
